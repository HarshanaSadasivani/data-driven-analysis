```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Here loading the dataset
file_path = "Disease_symptom_and_patient_profile_dataset.csv"
df = pd.read_csv(file_path)

print("First few rows of the dataset:")
print(df.head())

# Diaplaying the Summary statistics
print("\nSummary-statistics of the dataset:")
print(df.describe())

# Handling the data types and missing values
print("\nData types and the missing values:")
print(df.info())

# Target variable distribution
print("\nDistribution of the target variable:")
print(df['Outcome Variable'].value_counts())

# Visualization of the distribution of the target variable
plt.figure(figsize=(9, 7))
sns.countplot(x='Outcome Variable', data=df)
plt.title('Distribution of the Outcome Variable')
plt.xlabel('Outcome Variable')
plt.ylabel('Count')
plt.show()

# Visualizing the correlation matrix by excluding non-numeric columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(9, 7))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm', fmt=".3f")
plt.title('Correlation Matrix')
plt.show()
```
